```python
def transform_time(t1,t2):
    t1= str(t1)
    t2= str(t2)
    print(t1,t2)
    if(":" not in t1 or ":" not in t2):
        return float('NaN')
    else:
        t1h = int(t1.split(":")[0])*60
        t1m = int(t1.split(":")[1][0:2])
        t2h = int(t2.split(":")[0])*60
        t2m = int(t2.split(":")[1][0:2])
        total_time1 = t1h+t1m
        total_time2 = t2h+t2m
        return float(total_time1 - total_time2)
```

```python
def transform_time_order_picked(t):
    t= str(t)
    print(t)
    if(":" not in t):
        return float('NaN')
    else:
        if(int(t.split(":")[1])>=60):
            t=t.replace(str(t.split(":")[1]),"00")
            t=t.replace(t.split(":")[0],str(int(t.split(":")[0])+1))
        elif(int(t.split(":")[0])>=24):
            t=t.replace(str(t.split(":")[0]),"0")
            
        return t
```
